```python
import math

def qe_disc(a, b, c):
    # 判別式 D = b^2 - 4ac
    return b**2 - 4*a*c

def qe_solution1(a, b, c):
    # 判別式を計算
    d = qe_disc(a, b, c)
    # 解のうち大きくない方 (小さい方)
    return (-b - math.sqrt(d)) / (2 * a)

def qe_solution2(a, b, c):
    # 判別式を計算
    d = qe_disc(a, b, c)
    # 解のうち小さくない方 (大きい方)
    return (-b + math.sqrt(d)) / (2 * a)
```
