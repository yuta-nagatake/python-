```python
### 丸括弧で括る方法
x, y, z = (-1, -2, -3)
if (x < 0 and y < 0 and z < 0 and
    x != y and y != z and x != z):
    print('x, y and z are different and negatives.')

### 行末にバックスラッシュ(\) を入れる方法

x, y, z = (-1, -2, -3)
if x < 0 and y < 0 and z < 0 and \
    x != y and y != z and x != z:
    print('x, y and z are different and negatives.')
```
