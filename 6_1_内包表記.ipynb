```python
def allsubstrings(s):
  return [s[i:j] for i in range(len(s)) for j in range(i+1,len(s)+1)]
allsubstrings("abc")
```

```python
def var(lst):
    n = len(lst)
    av = sum(lst)/n
    return sum([(x - av)*(x - av) for x in lst])/n
```
