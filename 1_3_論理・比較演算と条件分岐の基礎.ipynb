```python
def absolute(x):
    if x < 0:
        return -x
    else:
        return x
```

```python
 absolute(-5)
```

```python
def sign(x):
    if x < 0:
        return -1
    if x > 0:
        return 1
    return 0
```

```python
sign(5)
```

```python
sign(0)
```

```python
#再帰
def fib(n):
    if n < 2:
        return n
    else:
        return fib(n-1) + fib(n-2)
```

```python
fib(10)
```
